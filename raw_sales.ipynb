```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv(r'D:\Datasets\raw_sales.csv')
```

```python
# Raw data
df
```

```python
df['datesold'] = pd.to_datetime(df['datesold'])
```

# 1 - Which date corresponds to the highest number of sales ?

```python
df['month'] = df['datesold'].dt.strftime('%Y-%m')
```

```python
date = df.groupby(['month'],as_index=False)['price'].sum()
highest_sale = date.sort_values(by='price',ascending=False).reset_index(drop=True)
highest_sale
```

# 2 - Find out the postcode with the highest average price per sales ?

```python
postcode = df.groupby(['postcode'],as_index=False)['price'].mean()
pd.options.display.float_format = "{:,.0f}".format
high_average_per_sales = postcode.sort_values(by='price',ascending=False).reset_index(drop=True)
high_average_per_sales
```

# 3 - Which year witnessed the lowest number of sales ?

```python
df['Year'] = df['datesold'].dt.strftime('%Y')
```

```python
lowest_num_of_sales = df.groupby(['Year'],as_index=False)['price'].sum()
lowest_num_of_sales
```

```python
plt.figure(figsize=(12,12))
plt.bar(lowest_num_of_sales['Year'],lowest_num_of_sales['price'])
plt.title('number of sales in each year',fontsize=15)


plt.savefig('Number.png')

plt.show()
```

# 4 - Use the window function to deduce the top six postcodes by year's price .

```python
window = df.groupby(['Year','postcode'],as_index=False)['price'].sum()
def top():
    y_2007 = window.loc[window['Year']=='2007']
    y_2007 = y_2007.sort_values(by='price',ascending=False)
    y_2007 = y_2007.iloc[:6]
    # -----------------------------------------
    y_2008 = window.loc[window['Year']=='2008']
    y_2008 = y_2008.sort_values(by='price',ascending=False)
    y_2008 = y_2008.iloc[:6]
    # -----------------------------------------
    y_2009 = window.loc[window['Year']=='2009']
    y_2009 = y_2009.sort_values(by='price',ascending=False)
    y_2009 = y_2009.iloc[:6]
    # -----------------------------------------
    y_2010 = window.loc[window['Year']=='2010']
    y_2010 = y_2010.sort_values(by='price',ascending=False)
    y_2010 = y_2010.iloc[:6]
    # -----------------------------------------
    y_2011 = window.loc[window['Year']=='2011']
    y_2011 = y_2011.sort_values(by='price',ascending=False)
    y_2011 = y_2011.iloc[:6]
    # -----------------------------------------
    y_2012 = window.loc[window['Year']=='2012']
    y_2012 = y_2012.sort_values(by='price',ascending=False)
    y_2012 = y_2012.iloc[:6]
    # -----------------------------------------
    y_2013 = window.loc[window['Year']=='2013']
    y_2013 = y_2013.sort_values(by='price',ascending=False)
    y_2013 = y_2013.iloc[:6]
    # -----------------------------------------
    y_2014 = window.loc[window['Year']=='2014']
    y_2014 = y_2014.sort_values(by='price',ascending=False)
    y_2014 = y_2014.iloc[:6]
    # -----------------------------------------
    y_2015 = window.loc[window['Year']=='2015']
    y_2015 = y_2015.sort_values(by='price',ascending=False)
    y_2015 = y_2015.iloc[:6]
    # -----------------------------------------
    y_2016 = window.loc[window['Year']=='2016']
    y_2016 = y_2016.sort_values(by='price',ascending=False)
    y_2016 = y_2016.iloc[:6]
    # -----------------------------------------
    y_2017 = window.loc[window['Year']=='2017']
    y_2017 = y_2017.sort_values(by='price',ascending=False)
    y_2017 = y_2017.iloc[:6]
    # -----------------------------------------
    y_2018 = window.loc[window['Year']=='2018']
    y_2018 = y_2018.sort_values(by='price',ascending=False)
    y_2018 = y_2018.iloc[:6]
    # -----------------------------------------
    y_2019 = window.loc[window['Year']=='2019']
    y_2019 = y_2019.sort_values(by='price',ascending=False)
    y_2019 = y_2019.iloc[:6]
    # -----------------------------------------
    top_6_postcode = pd.concat([y_2007,y_2008,y_2009,y_2010,y_2011,y_2012,y_2013,y_2014,y_2015,y_2016,y_2017,y_2018,y_2019])
    top_6_postcode['price'] = top_6_postcode['price'].astype(int)
    top_6_postcode.to_excel('top_6_postcode.xlsx',index=False)
    return top_6_postcode
top()


```
