# Mean of sales in specific range of areas 

```python
def range_first():
    a = df.loc[df['Store_Area'].between(770,1065)]
    sum_a_customer =a['Daily_Customer_Count'].sum()
    mean_a_sales = a['Store_Sales'].mean()
    mean_a_items = a['Items_Available'].mean()
    print('sum of customer in first range is : ',int(sum_a_customer))
    print('mean of sales in first range is : ',int(mean_a_sales))
    print('mean of items in first range is :',int(mean_a_items))
    
def range_second():
    b = df.loc[df['Store_Area'].between(1065,1360)]
    sum_b_customer =b['Daily_Customer_Count'].sum()
    mean_b_sales = b['Store_Sales'].mean()
    mean_b_items = b['Items_Available'].mean()
    print('sum of customer in second range is : ',int(sum_b_customer))
    print('mean of sales in second range is : ',int(mean_b_sales))
    print('mean of items in second range is :',int(mean_b_items))
    
    
def range_third():
    c = df.loc[df['Store_Area'].between(1360,1655)]
    sum_c_customer =c['Daily_Customer_Count'].sum()
    mean_c_sales = c['Store_Sales'].mean()
    mean_c_items = c['Items_Available'].mean()
    print('sum of customer in third range is : ',int(sum_c_customer))
    print('mean of sales in third range is : ',int(mean_c_sales))
    print('mean of items in third range is :',int(mean_c_items))
    
    
def range_fourth():
    d = df.loc[df['Store_Area'].between(1655,1950)]
    sum_d_customer =d['Daily_Customer_Count'].sum()
    mean_d_sales = d['Store_Sales'].mean()
    mean_d_items = d['Items_Available'].mean()
    print('sum of customer in fourth range is : ',int(sum_d_customer))
    print('mean of sales in fourth range is : ',int(mean_d_sales))
    print('mean of items in fourth range is :',int(mean_d_items))

def range_fifth():
    e = df.loc[df['Store_Area'].between(1950,2245)]
    sum_e_customer =e['Daily_Customer_Count'].sum()
    mean_e_sales = e['Store_Sales'].mean()
    mean_e_items = e['Items_Available'].mean()
    print('sum of customer in fifth range is : ',int(sum_e_customer))
    print('mean of sales in fifth range is : ',int(mean_e_sales))
    print('mean of items in fifth range is :',int(mean_e_items))
    
    
    
range_first() 
print('----------------------')
range_second()
print('----------------------')
range_third()
print('----------------------')
range_fourth()
print('----------------------')
range_fifth()
```

```python
data = {'range_of_area ( yard_square )' : ['[770 - 1065]','[1065 - 1360]','[1360 - 1655]','[1655 - 1950]','[1950 - 2245]']
        ,'mean_of_items' : [1163 ,1478,1800,2136,2432]
        ,'sum_of_customers' : [30710,186380,316780,156080,17920]
       ,'mean_of_sales' : [52951 , 58644 , 59273 , 60787 , 66659]}
```

```python
new_df = pd.DataFrame(data)
new_df
```

```python
plt.figure(figsize=(25,8))

plt.subplot(1,3,1)
plt.bar(new_df['range_of_area ( yard_square )'],new_df['mean_of_items'])
plt.xlabel('range_of_area ( yard_square )',fontsize=12)
plt.title('mean of items_available for each range of area',fontsize=14)

plt.subplot(1,3,2)
plt.bar(new_df['range_of_area ( yard_square )'],new_df['sum_of_customers'])
plt.xlabel('range_of_area ( yard_square )',fontsize=12)
plt.title('sum of customers for each range of area',fontsize=14)


plt.subplot(1,3,3)
plt.bar(new_df['range_of_area ( yard_square )'],new_df['mean_of_sales'])
plt.xlabel('range_of_area ( yard_square )',fontsize=12)
plt.title('mean of store_sales for each range of area',fontsize=14)


plt.savefig('Result.png',dpi=200)
plt.show()
```
