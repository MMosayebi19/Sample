```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime 
```

```python
# Open = Price when the market opens
# High = Highest recorded price for the day
# Low = Lowest recorded price for the day
# Close = Price when the market closes
# Adj Close = Modified closing price based on corporate actions
# Volume = Amount of stocks sold in a day
```

```python
df = pd.read_csv('D:\\Datasets\Tesla-dataset.csv')
```

```python
# Raw data
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['year'] = df['Date'].dt.strftime('%Y')
df['month'] = df['Date'].dt.strftime('%m')
df['day'] = df['Date'].dt.strftime('%d')
```

```python
# Transformed data
df
```

# Which month in each year we had highest Amount of stocks that sold in a day ?

```python
months = df.groupby(['month','year'],as_index=False)['Volume'].sum()
highest_amount = months.sort_values(by='year',ascending = False).reset_index(drop=True)
highest_amount['year'] = highest_amount['year'].astype(int)
highest_amount
```

```python
#highest_amount.to_excel('highest amount.xlsx',index=False)
```

```python
plt.figure(figsize=(20,20))
plt.suptitle('highest amount of a month per year',fontsize=20)

plt.subplot(231)
loc1 = highest_amount.loc[highest_amount['year']==2017]
loc1 = loc1.sort_values(by='month',ascending=True)
plt.plot(loc1['month'] , loc1['Volume'] , marker='o')
plt.xticks([i for i in range(0,13)])
plt.xlabel('Month')
plt.title('year = 2017')


plt.subplot(232)
loc2 = highest_amount.loc[highest_amount['year']==2018]
loc2 = loc2.sort_values(by='month',ascending=True)
plt.plot(loc2['month'] , loc2['Volume'] , marker='o')
plt.xlabel('Month')
plt.title('year = 2018')

plt.subplot(233)
loc3 = highest_amount.loc[highest_amount['year']==2019]
loc3 = loc3.sort_values(by='month',ascending=True)
plt.plot(loc3['month'] , loc3['Volume'] , marker='o')
plt.xlabel('Month')
plt.title('year = 2019')

plt.subplot(234)
loc4 = highest_amount.loc[highest_amount['year']==2020]
loc4 = loc4.sort_values(by='month',ascending=True)
plt.plot(loc4['month'] , loc4['Volume'] , marker='o')
plt.xlabel('Month')
plt.title('year = 2020')

plt.subplot(235)
loc5 = highest_amount.loc[highest_amount['year']==2021]
loc5 = loc5.sort_values(by='month',ascending=True)
plt.plot(loc5['month'] , loc5['Volume'] , marker='o')
plt.xlabel('Month')
plt.title('year = 2021')

plt.subplot(236)
loc6 = highest_amount.loc[highest_amount['year']==2022]
loc6 = loc6.sort_values(by='month',ascending=True)
plt.plot(loc6['month'] , loc6['Volume'] , marker='o')
plt.xlabel('Month')
plt.title('year = 2022')

plt.show()
```
